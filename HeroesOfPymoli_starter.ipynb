```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df= pd.read_csv(file_to_load)
#purchase_data_df.head(5)
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data_df["SN"].unique())
#type(player_count)

summary_table = pd.DataFrame({"Total Players": [player_count]})
summary_table
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = len(purchase_data_df["Item ID"].unique())

average_price = purchase_data_df["Price"].mean()
number_of_purchases = purchase_data_df["Item ID"].count()
total_revenue = purchase_data_df["Price"].sum()

summary_table = pd.DataFrame({"Number Of Unique Items":[unique_item],
                            "Average Price": [average_price],
                            "Number of Purchases": [number_of_purchases],
                            "Total Revenue":[total_revenue]})

summary_table["Average Price"] = summary_table["Average Price"].map("${:.2f}".format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].map("${:.2f}".format)                                                                    

summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_gender = purchase_data_df["Gender"].unique()

purchase_groups = purchase_data_df.groupby("SN").first()

gender_groups = purchase_groups["Gender"].value_counts()
                         
count_of_male = purchase_groups["Gender"].value_counts()["Male"]
count_of_female = purchase_groups["Gender"].value_counts()["Female"]
count_of_other = purchase_groups["Gender"].value_counts()["Other / Non-Disclosed"]

total_players = count_of_male + count_of_female + count_of_other

male_percent = count_of_male/total_players * 100
female_percent = count_of_female/total_players * 100
other_percent = count_of_other/total_players *100


gender_list = purchase_groups["Gender"].unique()
gender_count_list = (count_of_male,count_of_female, count_of_other)
percent_list = (male_percent,female_percent,other_percent)


summary_table = pd.DataFrame({"Gender": (gender_list),
                            "Percentage of Players":(percent_list),
                            "Total Counts" : (gender_count_list)})
                            

summary_table["Percentage of Players"] = summary_table["Percentage of Players"].map("{:.2f}".format)
 
                             
summary_table

                             


```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_gender_df = purchase_data_df.groupby("Gender")

purchase_data_count = purchase_gender_df["Purchase ID"].count()
purchase_data_avg = purchase_gender_df["Price"].mean()
purchase_data_sum = purchase_gender_df["Price"].sum()

purchase_unique_df = purchase_data_df.groupby("SN").first()
purchase_gender_unique = purchase_unique_df["Gender"].value_counts()

total_purchase_person = purchase_data_sum/purchase_gender_unique


summary_table = pd.DataFrame({"Purchase Count": (purchase_data_count),
                            "Average Purchase Price" : (purchase_data_avg),
                            "Average Purchase Value":(purchase_data_sum),
                            "Avg Purchase Total per Person" : (total_purchase_person)})



summary_table["Average Purchase Price"] = summary_table["Average Purchase Price"].map("${:.2f}".format)
summary_table["Avg Purchase Total per Person"] = summary_table["Avg Purchase Total per Person"].map("${:.2f}".format)
summary_table["Average Purchase Value"] = summary_table["Average Purchase Value"].map("${:.2f}".format)

summary_table
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_df = purchase_data_df.groupby("SN").first()

b_max = purchase_data_df["Age"].max()

bins = [0, 9, 14, 19, 24, 29, 34, 39, b_max]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39", "40+"]
                           
age_df["Age Group"] = pd.cut(age_df["Age"], bins, labels=group_names)

age_groups_df = age_df.groupby("Age Group")
total_ages = len(age_df["Age Group"])
age_count = age_groups_df["Age Group"].count()
age_percent = age_count/total_ages *100

summary_table = pd.DataFrame({"Percentage of Players" : (age_percent),
                              "Total Count" : (age_count)})
                            

summary_table["Percentage of Players"] = summary_table["Percentage of Players"].map("{:.2f}".format)
summary_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchase_df = purchase_data_df.groupby("SN").first()
pa_df = purchase_data_df

b_max = purchase_data_df["Age"].max()

bins = [0, 9, 14, 19, 24, 29, 34, 39, b_max]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39", "40+"]
                           
pa_df["Age Group"] = pd.cut(pa_df["Age"], bins, labels=group_names)
age_purchase_df["Age Group"] = pd.cut(age_purchase_df["Age"], bins, labels=group_names)

pa_index_df = pa_df.set_index("Age Group")
pa_index_groups_df = pa_index_df.groupby("Age Group")


total_items = pa_index_groups_df["Purchase ID"].count()
total_price= pa_index_groups_df["Price"].sum()
total_avg_purchase_price = pa_index_groups_df["Price"].mean()


avg_per_person_df= age_purchase_df.groupby("Age Group")
total_players = avg_per_person_df["Age Group"].count()

avg_per_person= total_price/total_players



summary_table = pd.DataFrame({"Purchase Count" : (total_items),
                            "Avgerage Purchase Price" : (total_avg_purchase_price),
                            "Total Purchase Value" : (total_price),
                           "Average Purchase Total per Person" : (avg_per_person)})



summary_table["Avgerage Purchase Price"] = summary_table["Avgerage Purchase Price"].map("${:.2f}".format)
summary_table["Total Purchase Value"] = summary_table["Total Purchase Value"].map("${:.2f}".format)
summary_table["Average Purchase Total per Person"] = summary_table["Average Purchase Total per Person"].map("${:.2f}".format)
summary_table


```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_df = purchase_data_df.groupby("SN")

SN_total_count = SN_df["SN"].count()
SN_avg_purchase = SN_df["Price"].mean()
SN_total_purchase = SN_df["Price"].sum()


summary_table = pd.DataFrame({"Purchase Count" : (SN_total_count),
                            "Average Purchase Price" : (SN_avg_purchase),
                            "Total Purchase Value" : (SN_total_purchase)})
                           
summary_table_sorted = summary_table.sort_values("Total Purchase Value", ascending=False)

    
summary_table_sorted["Average Purchase Price"] = summary_table_sorted["Average Purchase Price"].map("${:.2f}".format)
summary_table_sorted["Total Purchase Value"] = summary_table_sorted["Total Purchase Value"].map("${:.2f}".format)


summary_table_sorted
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
MPI_df= purchase_data_df.groupby(["Item ID", "Item Name"])

MPI_total_count = MPI_df["Item ID"].count()
MPI_avg_purchase = MPI_df["Price"].mean()
MPI_total_purchase = MPI_df["Price"].sum()

MPI_items_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
MPI_items_group = MPI_items_df.drop_duplicates()

Item_Price = MPI_items_group.sort_values("Item ID")
Item_Price_Index = Item_Price.set_index(["Item ID", "Item Name"])
                          
    
summary_table = pd.DataFrame({"Purchase Count": (MPI_total_count),
                            "Total Purchase Value" : (MPI_total_purchase)})
                     

merge_table = summary_table.merge(Item_Price_Index, how="outer", left_index= True, right_index= True)
MT_rename = merge_table.rename(columns= {"Price": "Item Price"})

summary_table_sorted = MT_rename.sort_values("Purchase Count", ascending=False)

summary_table_sorted ["Purchase Count"] = summary_table_sorted["Purchase Count"]
summary_table_sorted["Item Price"] = summary_table_sorted["Item Price"].map("${:.2f}".format)
summary_table_sorted["Total Purchase Value"] = summary_table_sorted["Total Purchase Value"].map("${:.2f}".format)


summary_table_final = summary_table_sorted[["Purchase Count", "Item Price", "Total Purchase Value"]]

summary_table_final
```
